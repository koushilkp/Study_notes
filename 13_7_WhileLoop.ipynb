## Factorial

```python
n = 5  # 5*4*3*2*1 =120
fact = 1
for i in range(1,n+1):
    fact*=i
fact
```

```python
n = 5
fact =1
i = n
while i>=1:# 5>=1(T),4>=1(T),3>=1(T)
    fact*=i # 1*5 =5,5*4=20,20*3=60
    i-=1   # 5-1=4,4-1=3,3-1=2
print(fact)
```
