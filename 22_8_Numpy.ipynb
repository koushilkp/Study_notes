## np.where()

```python
lst = [4,5,6,7,8,7]
index_lst = lst.index(7) # it return 1st occurence of index
index_lst
```

```python
np.where(condion) > return index value of an array element
return tuple
```

```python
lst = [4,5,6,7,8,8,3,4,5,6,7,8,5,46,78,8]
index_lst = []
for index, value in enumerate(lst):
    if value == 8:
        index_lst.append(index)
index_lst
```

```python
lst = [4,5,6,7,8,8,3,4,5,6,7,8,5,46,78,8]
index_lst = [index for index, value in enumerate(lst) if value == 8]
index_lst
```

```python
lst = [4,5,6,7,8,8,3,4,5,6,7,8,5,46,78,8]
arr = np.array(lst)
index_arr = np.where(arr==8)
index_arr
```

```python
lst = [4,5,6,7,8,8,3,4,5,6,7,8,5,46,78,8]
arr = np.array(lst)
index_arr = np.where(arr < 8) # 0 1 2 3 4 5 6 7 
index_arr
```

```python
lst = [4,5,6,7,8,8,3,4,5,6,7,8,5,46,78,8]
arr = np.array(lst)
index_arr = np.where(arr != 8) 
index_arr
```

## 3) std

How data value deviated from mean point/value
1std >>> Means 68% data are located near mean
Larger std means larger spread of data and vice versa
>> when we say data is normally distributed >>> when we have symmetricty about mean (50% data lie on left side and 50% will be on right side

```python
3.381-  1.118 
```

```python
let us consider height of peoples 3f,4f,5f,6f
> calculate mean and std >> 4.5 and 1.118
>> righ side >> 4.5 + 1.118 = 5.618,  5.618 + 1.118 = 6.736
>> left side >> 4.5 - 1.118 = 3.381 , 3.381-  1.118 = 2.263

1std >> means 68% of population /peoples between 3.381f to 5.618 f
2std >> means 95% of population /peoples between 2.263f to 6.736 f
```

```python
np.std([3,4,5,6])
```
