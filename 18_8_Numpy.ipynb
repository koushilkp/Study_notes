## sort()

```python
arr = np.random.randint(10,20,size=10)
print("Array\n",arr)
sort_array = np.sort(arr)
sort_array # ascending orde
```

```python
arr = np.random.randint(10,20,size=10)
print("Array\n",arr)
sort_array = np.sort(arr)[::-1]
sort_array # descending order
```

```python
arr = np.random.randint(10,20,size=10)
print("Array\n",arr)
sort_array = np.flip(np.sort(arr))
sort_array # Descending orde
```

## 2D array()

```python
arr  = np.array([[2,5,1],
                [5,0,1]])
print("Array\n",arr)
sort_arr = np.sort(arr)
sort_arr  # Ascending order > row wise
```

```python
arr  = np.array([[2,5,1],
                [5,0,1]])
print("Array\n",arr)
sort_arr = np.sort(arr)[::1,::-1]
sort_arr  # Descending order > row wise
```

```python
arr  = np.array([[2,5,1],
                [5,0,1]])
print("Array\n",arr)
sort_arr = -np.sort(-arr)
sort_arr  # Descending order > row wise
```

```python
arr  = np.array([[3,2,4],
                [5,0,1]])
print("Array\n",arr)
sort_arr = np.sort(arr)
sort_arr  # ascending order > Row wise
```

```python
arr  = np.array([[3,2,4],
                [5,0,1]])
print("Array\n",arr)
sort_arr = -np.sort(-arr)
sort_arr  # Descending order > Row wise
```

```python
arr  = np.array([[3,2,4],
                [5,0,1]])
print("Array\n",arr)
sort_arr = np.sort(arr)
sort_arr  # Descending order > col wise
```

```python
array([[2, 3, 4],
       [0, 1, 5]]) Ascending row

rray([[4, 3, 2],
       [5, 1, 0]])  Descending  row

array([[5, 2, 4],
       [3, 0, 1]])  Descending  col
array([[3, 0, 1],
       [5, 2, 4]]) # Ascending order > col wise
```

```python
arr  = np.array([[3,2,4],
                [5,0,1]])
print("Array\n",arr)
sort_arr = -np.sort(-arr,axis=0)
sort_arr  # Descending order > col wise
```

```python
arr  = np.array([[3,2,4],
                [5,0,1]])
print("Array\n",arr)
sort_arr = np.sort(arr,axis=0)
sort_arr  # Ascending order > col wise
```
